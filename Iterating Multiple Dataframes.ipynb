```python
import pandas as pd 
```

```python
product=pd.read_excel("/Users/arjodh_singh/Desktop/Test.xlsx",sheet_name='dim')
```

```python
sales=pd.read_excel("/Users/arjodh_singh/Desktop/Test.xlsx",sheet_name='fact')
```

# Iterating over mutiple data frames consectively

```python
for index,row in product.iterrows():
    m=sales[sales['pid']==row['pid']]['units'].to_dict()
    for i,j in m.items():
        sales.loc[i,'new']=m[i]*row['price']
```

```python
#sales
```

# Merging of two dataframes

```python
sales=pd.merge(product,sales,how='left',on='pid')
```

```python
sales['new2']=sales['price']*sales['units']
sales
```

# Transform the group by 

```python
sales['total']=sales.groupby(by='pid')['new'].transform('sum')
```

```python
sales['per%']=round((sales['new2']/sales['total'])*100,2)
```
